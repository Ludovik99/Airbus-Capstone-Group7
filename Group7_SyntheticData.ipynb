# Reduced Dataset

Creating a reduced Dataset with the derivative values of Fuel on Board, Total Consumption (aggregation of the 7 fuel quantity columns) and Flight number

```python
df_reduced = data[['Dx_VALUE_FOB', 'TOTAL_CONSUMED', 'Flight']]
```

```python
df_reduced
```

```python
df_reduced['Dx_VALUE_FOB'].mean()
```

```python
# Number of unique flights
df_reduced['Flight'].nunique()
```

```python
# Mean derivative value of FOB for the dataset
mean_fob = df_reduced['Dx_VALUE_FOB'].mean()
# Mean derivative value of Total Consumption for the dataset
mean_consumed = df_reduced['TOTAL_CONSUMED'].mean()
flight = 2025

y = 300   # Number of rows with a varying value
x = 11116 - y  # Number of rows to add
z = 50   # Multiplier for varying value

new_rows = []
varying_group = []

for i in range(x):
    row = [mean_fob, mean_consumed, flight]  # Initializing with constant values
    if i < y:
        row[0] = row[0] * z  # Setting varying value as z times the constant value
        row[1] = row[1] * z
        varying_group.append(row)
    new_rows.append(row)

# Find the position where the varying group should be inserted
insert_index = len(new_rows) - x//2

leak_flight = new_rows[:insert_index] + varying_group + new_rows[insert_index:]

leak_flight_df = pd.DataFrame(leak_flight, columns=['Dx_VALUE_FOB', 'TOTAL_CONSUMED', 'Flight'])

synthetic_reduced = pd.concat([df_reduced, leak_flight_df], axis=0)

synthetic_reduced = synthetic_reduced.reset_index(drop=True)

synthetic_reduced.iloc[3718224]
```

```python
# Mean values for each column
mean_values = {}

for column in df_regular.columns:
    mean_values[f"mean_{column}"] = df_regular[column].mean()

y = 30   # Number of rows with a varying value
x = 11116 - y  # Number of rows to add
z = 5   # Multiplier for varying value
flight = 2026

new_rows = []
varying_group = []

for i in range(x):
    row = [mean_values['mean_Dx_VALUE_FOB'], mean_values['mean_Dx_VALUE_FUEL_QTY_CT'], mean_values['mean_Dx_VALUE_FUEL_QTY_FT1'], mean_values['mean_Dx_VALUE_FUEL_QTY_FT2'],mean_values['mean_Dx_VALUE_FUEL_QTY_FT3'], mean_values['mean_Dx_VALUE_FUEL_QTY_FT4'], mean_values['mean_Dx_VALUE_FUEL_QTY_LXT'], mean_values['mean_Dx_VALUE_FUEL_QTY_RXT'], flight]  # Initializing with constant values
    if i < y:
      for n in range(8):
        row[n] = row[n] * z  # Setting varying value as z times the constant value
        varying_group.append(row)
    new_rows.append(row)

# Find the position where the varying group should be inserted
insert_index = len(new_rows) - x//2

leak_flight = new_rows[:insert_index] + varying_group + new_rows[insert_index:]

leak_flight_df = pd.DataFrame(leak_flight, columns=['Dx_VALUE_FOB',	'Dx_VALUE_FUEL_QTY_CT',	'Dx_VALUE_FUEL_QTY_FT1',	'Dx_VALUE_FUEL_QTY_FT2',	'Dx_VALUE_FUEL_QTY_FT3',	'Dx_VALUE_FUEL_QTY_FT4',	'Dx_VALUE_FUEL_QTY_LXT',	'Dx_VALUE_FUEL_QTY_RXT',	'Flight'])

synthetic = pd.concat([df_regular, leak_flight_df], axis=0)

synthetic = synthetic.reset_index(drop=True)

synthetic
```
